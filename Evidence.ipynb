```python
import base64

def decode_base64(data):
    missing_padding = 4 - len(data) % 4
    if missing_padding:
        data += '=' * missing_padding
    return base64.b64decode(data).decode('utf-8')

print(decode_base64("YWRtaW4"))
print(decode_base64("cGljb0NURns1M3J2M3JfNTNydjNyXzUzcnYzcl81M3J2M3JfNTNydjNyfQ"))

#picoCTF{53rv3r_53rv3r_53rv3r_53rv3r_53rv3r}.
```

## Birthday

The description strongly implies that you need to find two strings that generate a collision with md5 hash, i tried brute forecing it but it was taking too long so i decided to use a known collision. Then i uploaded the two strings to the website as pdf files and got the flag.

```python
#i used a know collision i found in https://crypto.stackexchange.com/questions/1434/are-there-two-known-strings-which-have-the-same-md5-hash-value
s1="d131dd02c5e6eec4693d9a0698aff95c 2fcab58712467eab4004583eb8fb7f89 55ad340609f4b30283e488832571415a 085125e8f7cdc99fd91dbdf280373c5b d8823e3156348f5bae6dacd436c919c6 dd53e2b487da03fd02396306d248cda0 e99f33420f577ee8ce54b67080a80d1e c69821bcb6a8839396f9652b6ff72a70"
s2="d131dd02c5e6eec4693d9a0698aff95c 2fcab50712467eab4004583eb8fb7f89 55ad340609f4b30283e4888325f1415a 085125e8f7cdc99fd91dbd7280373c5b d8823e3156348f5bae6dacd436c919c6 dd53e23487da03fd02396306d248cda0 e99f33420f577ee8ce54b67080280d1e c69821bcb6a8839396f965ab6ff72a70"

#we verify that the strings are not quite the same
for i in range(len(s1)):
    if s1[i]!=s2[i]:
        print(i)

#we save each string in its own pdf file 
f=open("s1.pdf","wb")
f.write(bytes.fromhex(s1.replace(" ","")))
f.close()
f=open("s2.pdf","wb")
f.write(bytes.fromhex(s2.replace(" ","")))
f.close()

```
